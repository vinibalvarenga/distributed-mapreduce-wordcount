```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura dos dados
df = pd.read_csv('results.csv')

# Calcular a razão (Communication Time + Synchronization Time) / Computation Time
df['ratio'] = (df['Communication Time'] + df['Synchronization Time']) / df['Computation Time']

# Visualizar os dados
print(df)

# Gráficos
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plotando os tempos individuais
axs[0, 0].plot(df['Number of Servers'], df['Communication Time'], marker='o', label='Communication Time')
axs[0, 0].plot(df['Number of Servers'], df['Computation Time'], marker='o', label='Computation Time')
axs[0, 0].plot(df['Number of Servers'], df['Synchronization Time'], marker='o', label='Synchronization Time')
axs[0, 0].set_title('Time Metrics vs Number of Servers')
axs[0, 0].set_xlabel('Number of Servers')
axs[0, 0].set_ylabel('Time (ns)')
axs[0, 0].legend()

# Plotando a razão
axs[0, 1].plot(df['Number of Servers'], df['ratio'], marker='o', color='purple')
axs[0, 1].set_title('Ratio (Comm Time + Sync Time) / Computation Time')
axs[0, 1].set_xlabel('Number of Servers')
axs[0, 1].set_ylabel('Ratio')

# Tempo total
df['Total Time'] = df['Communication Time'] + df['Computation Time'] + df['Synchronization Time']
axs[1, 0].plot(df['Number of Servers'], df['Total Time'], marker='o', color='green')
axs[1, 0].set_title('Total Time vs Number of Servers')
axs[1, 0].set_xlabel('Number of Servers')
axs[1, 0].set_ylabel('Total Time (ns)')

# Speedup
initial_computation_time = df['Computation Time'][0]
df['Speedup'] = initial_computation_time / df['Computation Time']
axs[1, 1].plot(df['Number of Servers'], df['Speedup'], marker='o', color='red')
axs[1, 1].set_title('Speedup vs Number of Servers')
axs[1, 1].set_xlabel('Number of Servers')
axs[1, 1].set_ylabel('Speedup')

plt.tight_layout()
plt.show()
```
